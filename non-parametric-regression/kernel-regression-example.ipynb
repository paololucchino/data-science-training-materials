```python
from statsmodels.nonparametric.kernel_regression import KernelReg
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Create some fake data using a polynomial function
p3 = .0005
p2 = 0
p1 = -2.5
p0 = 10
```

```python
x = np.linspace(-100,100,200)
y = p3*x**3 + p2*x**2 + p1*x + p0
data = y + np.random.normal(0,50,size=x.shape)
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,y, label='True relatonship')
plt.plot(x,data,'ro', label='Data')
plt.legend()
plt.show()
```

```python
kernel_reg = KernelReg(data, x, var_type='c', reg_type='lc', bw='cv_ls')
```

```python
y_pred, _ = kernel_reg.fit(x)
y_pred.shape
```

```python
plt.figure(figsize=(10,7))
plt.plot(x,y, label='True relatonship')
plt.plot(x,y_pred, label='Estimated relationship')
plt.plot(x,data,'ro', label='Data')
plt.legend()
plt.show()
```
